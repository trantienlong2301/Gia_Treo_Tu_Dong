```python
a,th1 = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_TRIANGLE)
plt.imshow(th1,cmap='gray')
th1.shape
```

```python
img1=np.zeros(th1.shape,dtype=np.uint8)

plt.imshow(img1,'gray')
img1.shape
```

```python
img2=cv2.circle(img1,(x,y),350,(255,255,255),-1)

plt.imshow(img2,'gray')
```

```python
result= cv2.bitwise_and(img2,th1)
```

```python
plt.imshow(result,'gray')
```

```python
def creat_line (x):
    img=np.zeros(th1.shape,dtype=np.uint8)
    angle_radian=np.deg2rad(x)
    end_point_x= int(480 - 440*np.cos(angle_radian))
    end_point_y=int(1280 - 440*np.sin(angle_radian))
    img_line = cv2.line(img, (480,1280),(end_point_x,end_point_y),(255,255,255),5)
    return img_line
```

```python
img_30=creat_line(30)
img_60=creat_line(60)
img_90=creat_line(90)
img_120=creat_line(120)
img_150=creat_line(150)
```

```python
result_30=cv2.bitwise_and(result,img_30)
result_60=cv2.bitwise_and(result,img_60)
result_90=cv2.bitwise_and(result,img_90)
result_120=cv2.bitwise_and(result,img_120)
result_150=cv2.bitwise_and(result,img_150)
```

```python
def count_pixel (img):
    return np.count_nonzero(img)
```

```python
max_value = 0
dem = 0
x1=count_pixel(result_30)
x2=count_pixel(result_60)
x3=count_pixel(result_90)
x4=count_pixel(result_120)
x5=count_pixel(result_150)
f =[x1,x2,x3,x4,x5]
for i in f:
    if i>max_value:
        max_value=i
        dem = dem+1

def vitri(x):
    if x == 1:
        return 30
    if x ==2:
        return 60
    if x== 3:
        return 90
    if x==4:
        return 120
    if x==5:
        return 150
vitri(dem)
```
